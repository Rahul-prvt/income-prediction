```python
from sklearn.feature_selection import mutual_info_classif

X = df.drop(columns=['Income'])  # Features
y = df['Income']

mi_scores = mutual_info_classif(X.select_dtypes(include=['int64', 'float64']), y)
mi_series = pd.Series(mi_scores, index=X.select_dtypes(include=['int64', 'float64']).columns)
mi_series.sort_values(ascending=False).plot(kind="bar", figsize=(10, 5), title="Mutual Information with Income")
```

```python
X = X.drop(columns=['fnlwgt','education-num','race','capital-loss'])
```

```python
cat_columns = X.select_dtypes('O')
```

```python
cat_columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
education = LabelEncoder()
marital_status = LabelEncoder()
occupation = LabelEncoder()
relationship = LabelEncoder()
native_country = LabelEncoder()
sex = LabelEncoder()
work_class = LabelEncoder()

# this variable store the encoded classes so this is need to dump 
```

```python
wc = X['education']
ed =X['marital-status']
ma = X['occupation']
re =X['relationship']
nc =X['native-country']
se =X['sex']
wc =X['workclass']
```

```python
eduencoded = education.fit_transform(wc)
marencoded = marital_status.fit_transform(ed)
occuencoded = occupation.fit_transform(ma)
relencoded = relationship.fit_transform(re)
natencoded = native_country.fit_transform(nc)
sexencoded = sex.fit_transform(se)
workencoded = work_class.fit_transform(wc)
```

```python
X['education'] = eduencoded
X['marital-status']= marencoded
X['occupation']= occuencoded
X['relationship'] =relencoded
X['native-country'] =natencoded
X['sex'] =sexencoded
X['workclass']=workencoded
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
xscaled = sc.fit_transform(X)
```

```python
xscaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(xscaled, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,f1_score
```

```python
ypred = model.predict(X_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
precision_score(y_test,ypred)
```

```python
confusion_matrix(y_test,ypred)
```

```python
f1_score(y_test,ypred)
```
